```python
import matplotlib.pyplot as plt

from maploc.osm.viz import Colormap
from vigor_osm_handler import get_osm_raster

%load_ext autoreload
%autoreload 2
```

# Example

`get_osm_raster((lat, long))` will give you back the tile

```python
raster = get_osm_raster(latlong=(46.5191, 6.5668))

map_viz = Colormap.apply(raster)
plt.imshow(map_viz)
```

You can also set the tile size (in meters) and the resolution by tweeking the pixel per meter parameter.

```python
raster = get_osm_raster(latlong=(46.5191, 6.5668), tile_size=800.0, ppm=5)

map_viz = Colormap.apply(raster)
plt.imshow(map_viz)
```

You have an example file in called `vigor_osm_handler.py` that downloads all the tiles for the dataset.
